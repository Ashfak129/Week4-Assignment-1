```python
Q1, Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed
and average_of_vehicle.
Ans:- Sure, here's an example of a Python class called Vehicle with an __init__ method that initializes instance 
variables name_of_vehicle, max_speed, and average_of_vehicle:
#Program
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle
    #I can create instances of this class and set the values for these instance variables like this:
car = Vehicle("Car", 200, 60)
bike = Vehicle("Bike", 150, 40)

print(car.name_of_vehicle)  # Output: Car
print(car.max_speed)        # Output: 200
print(car.average_of_vehicle)  # Output: 60

print(bike.name_of_vehicle)  # Output: Bike
print(bike.max_speed)        # Output: 150
print(bike.average_of_vehicle)  # Output: 40
```

```python
Q2. Create a child class car from the vehicle class created in Que 1, which will inherit the vehicle class.
Create a method named seating_capacity which takes capacity as an argument and returns the name of
the vehicle and its seating capacity.
Ans:- Here's an example of how can I create a child class Car that inherits from the Vehicle class and includes the
seating_capacity method:
#Program
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle

class Car(Vehicle):
    def seating_capacity(self, capacity):
        return f"{self.name_of_vehicle} has a seating capacity of {capacity} passengers."

# Creating instances of the Car class
car = Car("Car", 200, 60)
print(car.seating_capacity(5))  # Output: Car has a seating capacity of 5 passengers.

# I can also access the attributes inherited from the Vehicle class
print(car.max_speed)  # Output: 200
print(car.average_of_vehicle)  # Output: 60
```

```python
Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.
Ans:- Multiple inheritance is a feature in object-oriented programming languages that allows a class to inherit
attributes and methods from more than one parent class. In Python, a class can inherit from multiple parent classes,
and the child class will have access to the attributes and methods of all its parent classes.
    Here's an example to demonstrate multiple inheritance in Python:
#Program
class Parent1:
    def method1(self):
        print("This is method 1 from Parent 1")

class Parent2:
    def method2(self):
        print("This is method 2 from Parent 2")

class Child(Parent1, Parent2):
    def method3(self):
        print("This is method 3 from Child")

# Creating an instance of the Child class
child_instance = Child()

# Using methods from both parent classes and the child class
child_instance.method1()  # Output: This is method 1 from Parent 1
child_instance.method2()  # Output: This is method 2 from Parent 2
child_instance.method3()  # Output: This is method 3 from Child
```

```python
Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this
class.
Ans:- In Python, getters and setters are methods used to access and modify the attributes (instance variables) of a
class. They provide a way to control the access and modification of attributes, allowing you to encapsulate the internal
implementation details of a class.
    Here's an example class with getter and setter methods:
#Program
class Person:
    def __init__(self, name, age):
        self._name = name  # Convention: Use a single underscore to indicate a "protected" attribute
        self._age = age

    # Getter method for name
    def get_name(self):
        return self._name

    # Setter method for name
    def set_name(self, new_name):
        self._name = new_name

    # Getter method for age
    def get_age(self):
        return self._age

    # Setter method for age
    def set_age(self, new_age):
        if new_age >= 0:
            self._age = new_age
        else:
            print("Age cannot be negative!")

# Creating an instance of the Person class
person = Person("Alice", 30)

# Using getter and setter methods
print(person.get_name())  # Output: Alice
person.set_name("Bob")
print(person.get_name())  # Output: Bob

print(person.get_age())   # Output: 30
person.set_age(25)
print(person.get_age())   # Output: 25

person.set_age(-5)  # Output: Age cannot be negative!

```
